# Data Preprocessing

## Missing Data

```python
winedata.dropna(inplace=True)
winedata.head()
```

```python
print(winedata.shape)
```

## Normalization

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
X = winedata[['volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','alcohol']]
y = winedata[['quality']]

plt.figure()
plt.boxplot(X)
plt.xticks(np.arange(1, X.shape[1] + 1), winedata[['volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','alcohol']], rotation=30, ha="right")
plt.ylabel("quality")
plt.show()

X_m = MinMaxScaler().fit_transform(X)
plt.figure()
plt.boxplot(X_m)
plt.xticks(np.arange(1, X_m.shape[1] + 1), winedata[['volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','density','alcohol']], rotation=30, ha="right")
plt.ylabel("quality")
plt.show()
```

# 

# Neural Network

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_m, y, 
                                                    test_size = 0.25,
                                                    random_state = 123)
```

```python
from sklearn.neural_network import MLPRegressor

fit1 = MLPRegressor(random_state=529,
        hidden_layer_sizes=(400,200), activation='relu',solver='adam',
        alpha=0.0001,max_iter=300)

fit1.fit(X_train,y_train)
```

```python
NN_pred_test = fit1.predict(X_test)
NN_mse = mean_squared_error(NN_pred_test,y_test)
print ("Neural Network MSE =",NN_mse)
```

```python
NN_mae = mean_absolute_error(NN_pred_test,y_test)
print ("Neural Network MAE =", NN_mae)
```

# 

# Multivariate Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
MLR_pred_test = lr.predict(X_test)
MLR_mse = mean_squared_error(MLR_pred_test,y_test)
print ("Linear Regression MSE =",MLR_mse)
```

```python
MLR_mae = mean_absolute_error(MLR_pred_test,y_test)
print ("Linear Regression MAE =", MLR_mae)
```

```python
print("Intercept: ",lr.intercept_)
print("Coefficient: \n",lr.coef_)
```

# Ridge Regression

```python
from sklearn.linear_model import Ridge

RR = Ridge(alpha=3).fit(X_train,y_train)
```

```python
RR_pred_test = RR.predict(X_test)
RR_mse = mean_squared_error(RR_pred_test,y_test)
print ("Ridge Regression MSE =",RR_mse)
```

```python
RR_mae = mean_absolute_error(RR_pred_test,y_test)
print ("Ridge Regression MAE =", RR_mae)
```

```python
print("Intercept: ", RR.intercept_)
print("Coefficient: \n", RR.coef_)
```
